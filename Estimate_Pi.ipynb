```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np


def x(r, theta):
    return r*np.cos(theta)


def y(r, theta):
    return r*np.sin(theta)


theta = np.linspace(0, np.pi/2, 1001)
radius = 10
N = int(1e5)

x_array = x(r=radius, theta=theta)
y_array = y(r=radius, theta=theta)
x_rand_uni = np.random.uniform(0, radius, N)
y_rand_uni = np.random.uniform(0, radius, N)

x_inside = []
y_inside = []

for i in zip(x_rand_uni, y_rand_uni):
    if np.sqrt((i[0]*i[0]) + (i[1]*i[1])) <= radius:
        x_inside.append(i[0])
        y_inside.append(i[1])
    else:
        pass

inside = len(x_inside)
print('{0} out of {1} points inside circle.'.format(inside, N))

pi_est = (inside/N)*4
accuracy = pi_est/np.pi
print(
        'Estimate of Pi calculated to be {0} '
        'using a sample of {1} points.'.format(pi_est, N))
print(accuracy)

fig, ax = plt.subplots(1, 1)
fig.set_size_inches(8, 8)
ax.set_xlim(0, radius)
ax.set_ylim(0, radius)

ax.scatter(x_rand_uni, y_rand_uni, s=0.25)
ax.scatter(x_inside, y_inside, s=0.25, alpha=0.5)
ax.scatter(x_array, y_array)

plt.show()
```

```python
# Rosetta Code - https://rosettacode.org/wiki/Pi#Python

def calcPi():
    q, r, t, k, n, l = 1, 0, 1, 1, 3, 3
    while True:
        if 4*q+r-t < n*t:
            yield n
            nr = 10*(r-n*t)
            n  = ((10*(3*q+r))//t)-10*n
            q  *= 10
            r  = nr
        else:
            nr = (2*q+r)*l
            nn = (q*(7*k)+2+(r*l))//(t*l)
            q  *= k
            t  *= l
            l  += 2
            k += 1
            n  = nn
            r  = nr
 
import sys
pi_digits = calcPi()
i = 0
for d in pi_digits:
    sys.stdout.write(str(d))
    i += 1
    if i == 80: print(""); i = 0
```
